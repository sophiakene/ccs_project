```python
import pandas as pd
```

```python
df1 = pd.read_csv('../annotations/final_annotations/ALL_annotations.csv', index_col=0)
df2 = pd.read_csv('../annotations/random_without_party.csv', index_col=0)
df3 = pd.read_csv('../annotations/random_with_party.csv', index_col=0)
```

```python
df1
```

```python
del df2['created_time']
```

```python
df2['party'] = df3['party']
```

```python
df2 = df2.rename(columns={'message': 'comment_text'})
```

```python
df4 = pd.concat([df1, df2])
```

```python
drop_duplicates_df = df4.drop_duplicates(subset='comment_text')
```

```python
drop_duplicates_df
```

```python
drop_duplicates_df[drop_duplicates_df['comment_text'].str.contains(pat='http\S+')]['comment_text']
```

```python
drop_duplicates_df['comment_text'] = drop_duplicates_df['comment_text'].str.replace('http\S+', '')
```
