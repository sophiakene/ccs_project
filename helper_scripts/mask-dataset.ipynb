```python
import pandas as pd
```

```python
emoji_text_df = pd.read_csv('../dataset/text_emoji.csv', index_col=0)
```

```python
emoji_text_df
```

```python
import emoji
```

```python
no_emojis = [emoji.replace_emoji(emoji_text_df['comment_text'].iloc[i], '') for i in range(len(emoji_text_df))]
```

```python
text_df = pd.DataFrame(no_emojis, columns=['comment_text'])
```

```python
labels = list(emoji_text_df['label'])
```

```python
text_df['label'] = labels
```

```python
parties = list(emoji_text_df['party'])
```

```python
text_df['party'] = parties
```

```python
text_df.to_csv('../dataset/text.csv')
```

```python
only_emojis = []
for i in range(len(emoji_text_df)):
    emojis_dict = emoji.emoji_list(emoji_text_df['comment_text'].iloc[i])
    emojis_list = [emoji['emoji'] for emoji in emojis_dict]
    only_emojis.append(''.join(emojis_list))
```

```python
emojis_df = pd.DataFrame(only_emojis, columns=['comment_text'])
```

```python
emojis_df['label'] = labels
emojis_df['party'] = parties
```

```python
emojis_df.to_csv('../dataset/emoji.csv')
```

```python
text_df['label'] = labels
text_df['party'] = parties
```

```python
text_df = text_df[text_df['comment_text'] != '']
```

```python
text_df.to_csv('../dataset/text.csv')
```
